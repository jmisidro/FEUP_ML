```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sbn

# Load dataframes
awards_players = pd.read_csv('../data/awards_players.csv')
coaches = pd.read_csv('../data/coaches.csv')
players_teams = pd.read_csv('../data/players_teams.csv')
players = pd.read_csv('../data/players.csv')
series_post = pd.read_csv('../data/series_post.csv')
teams_post = pd.read_csv('../data/teams_post.csv')
teams = pd.read_csv('../data/teams.csv')
```

```python
plt.figure(figsize=(10, 6))
sbn.boxplot(x=awards_players['year'])
plt.title("Outliers in 'year' Column of awards_players Table")
plt.show()
```

```python
# Set up the plot
plt.figure(figsize=(12, 8))
sbn.stripplot(data=awards_players, x='year', y='award', jitter=True, dodge=True, marker='o', color='blue', alpha=0.7)

# Customize the plot
plt.title("Distribution of Awards Over Years")
plt.xlabel("Year")
plt.ylabel("Award Type")
plt.xticks(rotation=45)
plt.grid(True)

# Show the plot
plt.show()
```

```python



# plot the distribution of the rank in the teams table
plt.figure(figsize=(10, 6))
sbn.histplot(teams['rank'], bins=30, kde=True)
plt.title('Distribution of the rank in the teams table')
plt.xlabel('Rank')
plt.ylabel('Frequency')
plt.show()
```
